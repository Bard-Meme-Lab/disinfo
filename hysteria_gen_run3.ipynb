```
gpt2.download_gpt2(model_name="355M")
```

```
file_name = "hysteria.txt"
```

```
tf.reset_default_graph()
sess = gpt2.start_tf_sess()
gpt2.finetune(sess,
              dataset=file_name,
              model_name='355M',
              steps=900,
              restore_from='latest',
              run_name='run4',
              print_every=10,
              sample_every=200,
              save_every=500
              )
```

```
gpt2.mount_gdrive()
gpt2.copy_checkpoint_to_gdrive(run_name='run4')
```
